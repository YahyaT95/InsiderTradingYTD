```python
## REQUESTS_HTML #########################
# pip install requests-html
# pip install pandas

#This block of code is used to find the table of data within the html file. 
import requests_html
import pandas as pd

URL = 'http://openinsider.com/screener?s=&o=&pl=&ph=&ll=&lh=&fd=365&fdr=&td=0&tdr=&fdlyl=&fdlyh=&daysago=&xs=1&excludeDerivRelated=1&vl=50&vh=&ocl=&och=&sic1=-1&sicl=100&sich=9999&grp=0&nfl=&nfh=&nil=&nih=&nol=&noh=&v2l=&v2h=&oc2l=&oc2h=&sortcol=1&cnt=1000&page=1'

ses = requests_html.HTMLSession()
res = ses.get(URL)
html_table = res.html.find("table", first=True).html

df = pd.read_html(html_table)[11] # table found at index 11
```

```python
#reassign
sales_df=trade_df
```

```python
#Now we add new columns to clean them a
sales_df['delta_own']=sales_df['ΔOwn']
sales_df['delta_own']=sales_df['delta_own'].map(lambda x: x.lstrip('+->').rstrip(''))
sales_df['delta_own'] = sales_df['delta_own'].str.rstrip('%').astype('float') / 100.0

sales_df['trade_value']=sales_df['Value']
sales_df['trade_value']=sales_df['trade_value'].map(lambda x: x.lstrip('+-').rstrip(''))
sales_df['trade_value']=sales_df['trade_value'].str.replace('$', '')
sales_df['trade_value']=sales_df['trade_value'].str.replace(',', '')
sales_df['trade_value']=sales_df['trade_value'].astype('float')


#Could also use To 
#remove all substring "a" from column A: df["A"].str.replace("a","")
#To remove all substrings "a" or "b": df["A"].str.replace("[ab]","")
```

```python
sales_df['price_new']=sales_df['Price']
sales_df['price_new']=sales_df['price_new'].str.replace('$', '')
sales_df['price_new']=sales_df['price_new'].str.replace(',', '')
sales_df['price_new']=sales_df['price_new'].astype('float')
```

```python
#we can verify the types with a new column or just execute sales_df.dtypes
sales_df['delta_own_type'] = sales_df['delta_own'].apply(lambda x: type(x).__name__)
sales_df['trade_value_type'] = sales_df['trade_value'].apply(lambda x: type(x).__name__)
sales_df['price_new_type'] = sales_df['price_new'].apply(lambda x: type(x).__name__)
```

```python
#save sales_df for record 
x= 'insidersales_ytd_{}.csv'.format(pd.to_datetime("today").strftime("%m.%d.%Y"))
sales_df.to_csv(x, index=False)
```
